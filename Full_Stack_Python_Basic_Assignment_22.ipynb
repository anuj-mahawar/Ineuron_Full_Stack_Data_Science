**'X' is a global variable as any Python function/class can access this variable**

```python
X = 'iNeuron'
def func():
    X = 'NI!'
```

```python
func()
print(X)
```

**the scope of X inside the func is limited to the function itself. It serves as a local variable**  

```python
X = 'iNeuron'
def func():
    X = 'NI!'
    print(X)
```

```python
func()
print(X)
```

**the scope of X inside the func is limited to the function itself. It serves as a local variable.So, first print() is printed inside the func and then the one associated with global variable**

```python
X = 'iNeuron'
def func():
    global X
    X = 'NI!'
```

```python
func()
print(X)
```

**As X is defined global inside the function, it now serves as global variable and can be called outside the function**

```python
X = 'iNeuron'
def func():
    X = 'NI'
    def nested():
        print(X)
    nested()
```

```python
func()
```

```python
X
```

**X inside func is limited to the function itself and can't be accessed outside. So, the nested function has the access locally**


```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)
```

```python
func()
```

**nonlocal is to tell python explicitly that it has to access X initialized in nested on line 4**
